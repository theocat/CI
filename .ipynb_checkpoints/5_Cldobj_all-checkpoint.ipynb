```python
import os,struct,re,time
import numpy as np
import pandas as pd
```

```python
data_path = '/work/wuxue/data/CI/CI/'
```

```python
time_start = '20180614_0310'
time_end = '20180614_1000'
```

```python
time_s = time.mktime(time.strptime(time_start,'%Y%m%d_%H%M'))
time_e = time.mktime(time.strptime(time_end,'%Y%m%d_%H%M'))
times = np.arange(time_s,time_e+1,10*60)
```

```python
curdata = []
prepos_remark=[]
preID=[]
predata = []
[curdata.append('curdata_'+str(i)) for i in range(1,18)]
[prepos_remark.append('prepos_remark_'+str(i)) for i in range(1,51)]
[preID.append('preID_'+str(i)) for i in range(1,51)]
[predata.append('predata_'+str(i)) for i in range(1,18)]
Vars = ['curID','curcldsize']+ curdata +['curpos_lat_c','curpos_long_c','curAng_SunSeta','curAng_SatSeta','curAng_deltaPhi'] \
       + prepos_remark + preID +['precldsize']+ predata +['prepos_lat_c','prepos_long_c','preAng_SunSeta','preAng_SatSeta','preAng_deltaPhi']
Data = pd.DataFrame(columns=['Time']+Vars)
```

```python
for t in times:
    time_str = time.strftime("%Y%m%d_%H%M", time.localtime(t))
    print time_str
    filename = os.path.join(data_path, 'H8_'+time_str+'_CldRcrd_simple.dat')
    try:
        f = open(filename)
        s = f.read()
        s1 = re.split('\s+', s) #利用正则函数进行分割
        data_str = np.array(s1[1:-1]).reshape(-1,147)
        f.close()
    except:
        print 'file error: '+ filename
    data = np.zeros_like(data_str,dtype='float')
    data = data_str.astype('float')
    dD = pd.DataFrame(data,columns=Vars)
    dD['Time'] = time_str
    Data = pd.concat([Data, dD])
```

```python
Data[Data<-900000]=np.nan
```

```python
Data['CI'] = False
```
